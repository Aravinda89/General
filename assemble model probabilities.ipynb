```python
import pandas as pd
```

```python
df1 = pd.read_csv('submission (1).csv')
df1.head()
```

```python
df2 = pd.read_csv('submission_9.csv')
df2.head()
```

```python
df_asem_1 = pd.DataFrame(columns=['image_id','healthy','multiple_diseases','rust','scab'])
df_asem_1['image_id']=df1['image_id']
df_asem_1
```

```python
df_asem_1.iloc[:,1:] = df2.iloc[:,1:]*0.5 + df1.iloc[:,1:]*0.5
print(df_asem_1.head())
df_asem_1.to_csv('asem_1.csv',index=False)
```
